```python
import pandas as pd
import numpy as np
```

```python
directory = "/content/drive/MyDrive/Work & Courses/SAS Hackathon/hasil_sentiment/"
df = pd.read_csv(directory + "final_tweet_bener.csv")
```

```python
# prompt: count the number of null values in sentiment column of df

null_count = df['sentiment'].isnull().sum()
print(f"Number of null values in sentiment column: {null_count}")
```

```python
df.shape
df.head()
```

```python
df.columns
```

```python
# prompt: count the number of null values in sentiment column of df grouped by ruu and sorted descendingly

null_counts_by_ruu = df.groupby('ruu')['sentiment'].apply(lambda x: x.isnull().sum()).sort_values(ascending=False)
print(null_counts_by_ruu)
```

```python
df.columns
```

```python
columns = ['ruu'
          ,'user'
          ,'replying-to'
          ,'sentiment']
df_sliced = df[columns]
```

```python
df_sliced[['replying-to']] = df_sliced[['replying-to']].applymap(lambda x: eval(x, {'nan': np.nan}))
```

```python
df_explode = df_sliced.explode('replying-to')
df_explode.shape
```

```python
df_explode.to_csv(directory + 'final_tweet_bener_explode.csv',index=False)
```
